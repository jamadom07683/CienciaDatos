```python
# Convertir los títulos de las columnas a mayúsculas para secop1_df
secop1_df.columns = secop1_df.columns.str.upper()
```

```python
# Convertir los títulos de las columnas a mayúsculas para secop2_df
secop2_df.columns = secop2_df.columns.str.upper()
```

```python
#Generamos una funcion para una limpieza de datos que nos funcione para ambos dataframes

def limpiar_datos(df):
    df.columns = df.columns.str.strip()
    df = df.drop_duplicates()
    df = df.fillna('')
    # Convertir las cadenas de texto en las columnas a minúsculas
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.lower()

    # Eliminar caracteres especiales de las columnas tipo string
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.replace(r'[^\w\s]', '', regex=True)

    # Eliminar espacios en blanco adicionales en las columnas tipo string
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.strip()

    return df
```

```python
# Aplicar la función de limpieza a ambos DataFrames
secop1_df = limpiar_datos(secop1_df)
secop2_df = limpiar_datos(secop2_df)
```

```python
# Transformar 'anno_paa' en secop1_df a entero
secop1_df['ANNO_PAA'] = secop1_df['ANNO_PAA'].astype(int)
```

```python
# Transformar 'anno' en secop2_df a entero
secop2_df['ANNO'] = secop2_df['ANNO'].astype(int)
```

```python
#cruzaremos las tablas por una columna en comun, en este caso 'nit_entidad
cruce_df = pd.merge(secop1_df, secop2_df, left_on='NIT_ENTIDAD', right_on='NIT_ENTIDAD', how='inner')
```

```python
# Filtraemos secop1 por el año mayor igual a 2022
secop1_df_filtrado = secop1_df[secop1_df['ANNO_PAA'] >= 2022]
```

```python
#listamos el dataframe
display(secop1_df_filtrado)
```

```python
#Filtramos secop2 por Nit 
cruce_nit = pd.merge(secop1_df, secop2_df, how='inner', left_on='NIT_ENTIDAD', right_on='NIT_ENTIDAD')
```

```python
# Eliminar caracteres no numéricos de la columna NIT en secop1_df
secop1_df['NIT_ENTIDAD'] = secop1_df['NIT_ENTIDAD'].str.replace(r'\D', '', regex=True)
```

```python
# Eliminar caracteres no numéricos de la columna NIT en secop2_df
secop2_df['NIT_ENTIDAD'] = secop2_df['NIT_ENTIDAD'].str.replace(r'\D', '', regex=True)
```

```python
# Cruce final entre secop1_df y secop2_df basado en 'nit_entidad'
resultado_cruce_df = pd.merge(secop1_df, secop2_df, on='NIT_ENTIDAD', how='inner')
```

```python
#listamos el cruce
display(resultado_cruce_df)
```

```python
#cruzaremos ambas tablas por el nombre de la entidad
cruce_por_nombre = pd.merge(secop1_df, secop2_df, how='inner', on='NOMBRE_ENTIDAD')
```

```python
#importamos librerias para visualiacion
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Gráfico de Barras para `DEPARTAMENTO_PAA`
plt.figure(figsize=(12, 8))
secop1_df['DEPARTAMENTO_PAA'].value_counts().plot(kind='bar', color='lightgreen')
plt.xlabel('DEPARTAMENTO PAA')
plt.ylabel('Número de Registros')
plt.title('Número de Registros por Departamento PAA')
plt.xticks(rotation=90)
plt.show()
```

```python
#Graficos de registros por año
plt.figure(figsize=(10, 6))
secop1_df['ANNO_PAA'].value_counts().sort_index().plot(kind='bar', color='purple')
plt.xlabel('Año')
plt.ylabel('Número de Registros')
plt.title('Número de Registros por Año')
plt.grid(True)
plt.show()
```

```python
#Grafico de tipo pastel para visualizar la distribucion del "Estado" en secop2
plt.figure(figsize=(8, 8))
secop2_df['ESTADO'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Distribución del Estado')
plt.ylabel('')
plt.grid(True)
plt.show()
```
