#### Importando el Dataset

```python
#importar la libreria de Pandas
import pandas as pd
```

### Limpieza de un caracter especifico

```python
#limpieza caracter especial
empresas_df['CIUDAD DOMICILIO']=empresas_df['CIUDAD DOMICILIO'].str.replace(r'-.*$', '', regex=True) 
#nos situamos en la columna ciudad domicilio 
#utilizamos la funcion str replace, definimos 'r' para determinar que es texto, escribimos '-' que es el valor que estamos buscando, y luego '.*$' nos permite seleccionar todo a la derecha
#regextrue nos permite indicar que es una expresión regular 
```

##### Listamos nuevamente nuestro Dataframe y la ciudad deberia estar sola, sin departamento

```python
display(empresas_df)
```

#### Eliminar signo $ 

```python
# Eliminar el signo $ de todas las columnas
empresas_df = empresas_df.replace({r'\$': ''}, regex=True)
```

### Reduccion Nits longitud mayor a 9

```python
import numpy as np
#traemos una biblioteca de python para operaciones matematicas
```

```python
#Contamos cuantos valores de nit son mayores a 9
NitsMayores9=empresas_df['NIT'].astype(str).str.len() > 9
cuentaNits=np.sum(NitsMayores9)
if(cuentaNits==0): #Hacemos una sentencia condicional que verifica si existen nits con longitud mayor a 9
    print("NO SE TIENEN NITS CON LONGITUD MAYOR a 9") # de ser negativo, devuelve este mensaje
else:
    print("La cantidad de Nits con mas de caracteres es: " + cuentaNits) #de ser afirmativo, devuelve este mensaje
```

```python
#aplicaremos una funcion para reducir a maximo 9 los nit que tengan mas valores
empresas_df['NIT']=empresas_df['NIT'].astype(str).str.slice(0,9)
#Nos situamos en la columna de NIT, utilizamos la funcion astype(str) para convertir el tipo de dato a str, y luego utilizamos la funcion slice que extrae la columna indicandole el valor inciial y final, en este caso 0-9
```

```python
#Listamos la cantidad de nits con longitud mayor a 9
print("La cantidad de nits con longitud mayor a 9 es: " + str(cuentaNits))
```

### Transformando los tipos de datos

```python
#Creamos una lista para indicar a que columnas hacer el cambio
columns_to_convert = [
    'INGRESOS OPERACIONALES',
    'GANANCIA (PÉRDIDA)',
    'TOTAL ACTIVOS',
    'TOTAL PASIVOS',
    'TOTAL PATRIMONIO'
]
for column in columns_to_convert:
    # Eliminar caracteres no numéricos y convertir a float
    empresas_df[column] = empresas_df[column].replace({r'\$': '', ',': ''}, regex=True).astype(float)
display(empresas_df)
```

```python
#Definir tipo de variable por columna de nuestro dataset
datosTransformados=empresas_df.dtypes
print(datosTransformados)
```

```python
# Filtrar las 10 mejores empresas del año 2021 por TOTAL PATRIMONIO
top_empresas_2021 = empresas_df[empresas_df['Año de Corte'] == 2021].nlargest(10, 'TOTAL PATRIMONIO')
display(top_empresas_2021)
```

```python
# Filtrar las 10 mejores empresas del año 2021 por TOTAL PATRIMONIO
top_empresas_2022 = empresas_df[empresas_df['Año de Corte'] == 2022].nlargest(10, 'TOTAL PATRIMONIO')
display(top_empresas_2022)
```
