## Funciones

### Limpieza de Datos

```python
# Eliminar valores nulos
def eliminarNulos(titanic_df): #Creamos una funcion sobre el datamart creado
    titanicNulos_df={} #creamos un nuevo diccionario donde ya no se tengan valores nulos
    for columna in titanic_df.columns: #Hacemos un recorrido por columna en cada variable de nuestro df
        valores_limpios = []
        for valor in titanic_df[columna]: 
            if valor is not None:  # Verifica si el valor no es nulo
                valores_limpios.append(valor) #Agregamos a la tabla Valores_Limpios los valores que no son nulos
        titanicNulos_df[columna] = valores_limpios #Definimos que nuestro nuevo df tendra los valores limpios
    return pd.DataFrame(titanicNulos_df) #Traemos el nuevo df
```

```python
#Llamar a la funcion de eliminacion de Nulos
titanicNuevo=eliminarNulos(titanic_df)
print(titanicNuevo)
```

```python
#Listamos el dataframe en un formato visual mas agradable
display(titanicNuevo)
```

```python
#Listamos los tipos de los datos
tiposDatos=titanicNuevo.dtypes
print(tiposDatos)
```

```python
#limpiar los espacios en blanco de los nombres
titanicNuevo['Name'] = titanicNuevo['Name'].str.strip()
```

```python
#convertir la edad a un numero entero
titanicNuevo['Age'] = titanicNuevo['Age'].astype(int)
```

```python
#reemplazar valores en sexo (binario) para un mejor analisis
titanicNuevo['Sex'] = titanicNuevo['Sex'].map({'male': 0, 'female': 1})
```

```python
#Verificar y depurar registros duplicados
titanicNuevo.drop_duplicates(inplace=True)
```

```python
#Listamos nuevamente nuestro dataset
display(titanicNuevo)
```
